```python
from translate import Translator
import speech_recognition as sr
import gtts as ts
import playsound as ps
import os

recognizer = sr.Recognizer()

with sr.Microphone() as source:
    print("Speak Now")
    voice = recognizer.listen(source)
    try:
        text = recognizer.recognize_google(voice, language="hi")
        print("You said: " + text)
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
    except sr.RequestError as e:
        print("Could not request results from Google Speech Recognition service; {0}".format(e))

translator = Translator(to_lang="mr")
translation = translator.translate(text)
print("Translation: " + translation)

# Specify a different path to avoid permission issues
file_path = "output_audio/hello.mp3"

# Create directory if it doesn't exist
os.makedirs(os.path.dirname(file_path), exist_ok=True)

# Check if the file exists and delete it if necessary
if os.path.exists(file_path):
    os.remove(file_path)

converted_audio = ts.gTTS(translation, lang="mr")
converted_audio.save(file_path)

# Play the audio file
ps.playsound(file_path)

```
